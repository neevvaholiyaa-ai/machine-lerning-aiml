```python
plt.figure(figsize=(15,10))
for j,i in enumerate(df1.select_dtypes(include='number').drop(columns=['%- change over previous year','%- change over the previous year','%- change over the previous year.1']).columns):
    plt.subplot(2,2,j+1)
    sns.histplot(df1[i],kde=True)
    plt.title(i)
plt.tight_layout()
plt.show()
```

<b>Year: </b>The distribution is left-skewed, with the majority of data concentrated between 2015 and 2020, indicating that tourist arrivals increased significantly after 2016.<br>

<b>FTAs in India (in million): </b>The distribution is right-skewed, with most values concentrated between 0 and 3 million, indicating that Foreign Tourist Arrivals (FTAs) in India have generally remained in the lower range across most years.<br>

<b>NRIs arrivals in India (in million): </b>The distribution is left-skewed, with most values concentrated between 6 and 6.5 million, indicating that NRI arrivals have remained fairly stable and consistent within this narrow range.<br>

<b>ITAs in India (in million): </b>The distribution is left-skewed, with most values concentrated from 13 million onward, indicating that Indian Tourist Arrivals (ITAs) have predominantly remained high in recent years.<br>

```python
plt.figure(figsize=(30,10))
plt.subplot(1,3,1)
sns.lineplot(data=df1,y=df1['FTAs in India (in million)'],x=df1['Year'])
plt.title("Relationship between FTA and Yar")
plt.subplot(1,3,2)
sns.lineplot(data=df1,y=df1['NRIs arrivals in India (in million)'],x=df1['Year'])
plt.title("Relationship between NRI and Yar")
plt.subplot(1,3,3)
sns.lineplot(data=df1,y=df1['ITAs in India  (in million)'],x=df1['Year'])
plt.title("Relationship between ITA and Yar")
plt.show()
```

<b>Overall Observation: </b>All three tourist categories (FTAs, NRIs, and ITAs) show a common pattern — steady growth until 2019, followed by a sharp decline in 2020 due to the COVID-19 pandemic. This confirms that 2020 was an anomalous year across all tourism metrics, and any future analysis/forecasting should treat it as an outlier year rather than a standard data point.<br>

```python
plt.figure(figsize=(15,10))
cols=df1.select_dtypes(include='number').corr()
sns.heatmap(cols,annot=True,cmap="Blues")
plt.title("heatmap")
plt.show() 
```

<b>Positive Relationship: </b>The 'Year' column shows a positive relationship with 'FTAs in India (in million)' — as the year increases, the number of FTAs in India also tends to increase.<br>

<b>Negative Relationship: </b>The 'Year' column shows a negative relationship with 'NRIs arrivals in India (in million)' — as the year increases, the number of NRI arrivals tends to decrease.<br>

<b>Independent Relationship: </b>The 'Year' column shows little to no relationship with '% Change over previous year' — indicating that 'Year' and '% Change over previous year' are largely independent of each other.<br>

<b>FTAs: </b>As usual, FTAs in India are increasing, which is a good sign. If we can strengthen our medical infrastructure, it will help minimize the impact of situations like COVID-19 or other viruses on FTA numbers in the future.<br>

<b>Distribution by Age-Group (in Year) - 0-14: </b>We can observe that the 0-14 age group is decreasing, likely due to travel safety concerns and personal reasons. If we provide better safety measures and free/discounted tickets for this group, there is a high chance of increasing this number.<br>

<b>Distribution by Age-Group (in Year) - 15-24: </b>This is a key group to focus on. From 2013 to 2019, this group showed a significant decline, so our aim should be to increase it. By developing school/college exchange programs and offering technical courses, we can attract more travelers in this age group, which will also benefit overall FTA numbers.<br>

<b>Distribution by Age-Group (in Year) - 25-34: </b>Compared to the 15-24 group, this group has also decreased, but not drastically — indicating high potential for growth. By providing government support for business setups and developing tech hubs, this group has a strong chance of increasing, as many in this age range travel to India for business purposes.<br>

<b>Distribution by Age-Group (in Year) - 35-44: </b>From 2016 to 2019, this group showed an increase, which is a positive sign. This group primarily travels for vacations and to meet relatives. By promoting India's tourist destinations and culture more widely, we can further increase this number.<br>

<b>Distribution by Age-Group (in Year) - 45-54: </b>This group's numbers have decreased from 2014 to 2019. Our aim should be to increase this by providing better safety, medical facilities, and high-quality services for this age group.<br>
